We must clean the dataset. Many variables are categorical and need dummy variables created. Some numerical variables have strings attatched to them. Some variables we should drop entirely.

To start, we can drop `id` and `member_id`. We should also drop the last two rows as they are not meaningful.

```python
df_a_raw = df_a_raw_copy.copy()
```

```python
df_a_clean = df_a_raw.drop(df_a_raw.index[[-1,-2]])
df_a_clean = df_a_clean.drop(['id', 'member_id'], axis=1)
```

```python
#statistical functions

#gets ratio of NaNs for each column
def stats_NaN(df):
    df_stats = pd.DataFrame(index=[df.columns], columns=["NaN Ratio"])
    for col in df.columns:
        df_stats["NaN Ratio"][col] = df[col].isna().sum()/len(df) #NaN ratio
    return df_stats.sort_values(by=['NaN Ratio'])

#helps decide what to make dummies for and what to drop
def stats_unique(df, labels):
    df_counts = pd.DataFrame(index=labels,columns=['Unique Count'])
    for l in labels:
        df_counts['Unique Count'][l] = df[l].nunique()
    return df_counts
```

```python
stats_nan = stats_NaN(df_a_clean)
```

We see that `emp_title` and `title` have too many values to create dummies for. So we should drop them. We should likely keep `addr_state` information as that could be relevant. Anything that has 2 unique values can be converted into a binary feature. Anything with more, we make dummies for. We do this for `term` as well because it can either be a short term loan or long term loan. The variables below are lists that correspond indicate the way these features need to be cleaned. 

```python
ratio_feats = ['int_rate','dti','dti_joint','bc_util','il_util','sec_app_revol_util']
date_feats = ['issue_d','earliest_cr_line','last_credit_pull_d','last_pymnt_d','next_pymnt_d']
binary_feats = ['pymnt_plan','application_type','disbursement_method','debt_settlement_flag',
                'initial_list_status']
fix_nan_inf_feats = ['mths_since_last_delinq','mths_since_last_record','mths_since_recent_bc_dlq',
                 'mths_since_recent_inq','mths_since_recent_revol_delinq']
drop_feats = ['url','desc', 'emp_title','title'] #other not useful features
dummy_feats = ['home_ownership','verification_status','loan_status','purpose','addr_state']
```

```python
#helpful cleaning functions
def convert_percent(df, labels):
    for label in labels:
        df[label] = df[label].apply(lambda s: float(s)/ 100.0)
        
def convert_date(df, labels, frmt='%b-%Y'):
    for label in labels:
        df[label] = df[label].apply(lambda s: pd.to_datetime(s, format=frmt))
        
def convert_binary(df, labels):
    def to_bin(x, valid):
        if not pd.isnull(x):
            if x == valid:
                return 1
            else:
                return 0
        return np.nan #keeps NaNs
            
    for label in labels:
        #get first non NaN value in column
        valid = df[label].loc[df[label].first_valid_index()]
        df[label] = df[label].apply(lambda x: to_bin(x, valid))
        
def replace_NaNs(df, labels, fill_val):
    for label in labels:
        df[label].fillna(fill_val, inplace=True)
        
def fix_term(s):
    if not pd.isnull(s):
        if int(s.strip()[0:2]) == 60:
            return 1
        else:
            return 0
    else:
        return np.nan

def fix_emp_length(x):
    if not pd.isnull(x):
        if '<' in x:
            return 0.5
        else:
            s = ''.join(l for l in x if l.isdigit())
            return float(s)
    else:
        return np.nan
```

```python
#functions to convert ordinal letter subgrade into a float and back
def subgrade_to_num(grade):
    if not pd.isnull(grade):
        letter = grade[0]
        num = int(grade[1])
        num_grade = {
            'A':0,
            'B':1,
            'C':2,
            'D':3,
            'E':4,
            'F':5,
            'G':6
        }.get(letter)
        num_grade += (num-1)*0.2
        return num_grade
    else:
        return np.nan
```

```python
large_mths = 800
def clean_df(df_raw, ratio_feats, date_feats, binary_feats, fix_nan_inf_feats, dummy_feats, drop_feats,
             term='term', emp_length='emp_length', sub_grade='sub_grade'):
    df = df_raw.copy() #use and return a fresh copy as to not overwrite raw dataset

    print("Starting clean...")
    convert_percent(df, ratio_feats)
    print("Converted percents...")
    convert_date(df, date_feats)
    print("Converted dates...")
    convert_binary(df, binary_feats)
    print("Converted binary features...")
    #replace_NaNs(df, fix_nan_inf_feats, np.infty) #Cant use infinity, proxy with 800
    replace_NaNs(df, fix_nan_inf_feats, large_mths)
    print("Replaced NaN with large value...")
    df = pd.get_dummies(df, columns=dummy_feats)
    print("Made dummies...")
    df.drop(drop_feats, axis=1, inplace=True)
    print("Dropped features...")
    df[term] = df[term].apply(fix_term)
    print("Fixed term feature...")
    df[emp_length]=df[emp_length].apply(fix_emp_length)
    print("Fixed emp_length feature...")
    df['num_grade'] = df[sub_grade].apply(subgrade_to_num)
    print("Made num_grade...")
    
    print("Done!")
    return df
```

```python
df_clean = clean_df(df_a_clean,ratio_feats,date_feats,binary_feats,fix_nan_inf_feats,dummy_feats,drop_feats)
```
